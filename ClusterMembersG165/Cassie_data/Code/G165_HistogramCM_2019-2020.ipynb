### This code will use the data in G165_cm_Cass2020_bino and G165_cm_Frye2019a.cat and make a histogram of the data

```python
# Load Modules
import numpy as np
import astropy.units as u
import statistics as stat

# import plotting modules
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
```

```python
# Import the files
filename1="./G165_combo_Frye_Cass.txt"
filename2="./G165_zrange_342_355.txt"
filename3="./G165_zrange_3sig.txt"


data1 = np.genfromtxt(filename1,dtype=None,names=True,skip_header=2)
data2 = np.genfromtxt(filename2,dtype=None,names=True,skip_header=2)
data2 = np.genfromtxt(filename3,dtype=None,names=True,skip_header=2)
```

```python
#plot all the candidates
plt.figure(figsize=(20,15))
ax = plt.subplot(111)
plt.hist(data1['z_spec'], color= 'brown', bins=np.arange(0.18,0.53,0.01), range=None, histtype='bar', align='mid',rwidth=.95, label ='all data in range')

#plot piece we will look at more closely
plt.hist(data1['z_spec'], color='red', bins=np.arange(0.33,0.37,0.01), range=None, histtype='bar', align='mid',rwidth=.95, label='possible members')

#labels
plt.title('Cluster Members per Redshift', fontsize=20)
plt.xlabel('redshift', fontsize=20)
plt.xticks(np.arange(0.18,0.53,0.02), fontsize=15)
plt.yticks(np.arange(0,26,2.5), fontsize=15)

legend = ax.legend(loc='upper left',fontsize='x-large')
```

```python
#plot the subset of candidates selected on previous plot
plt.figure(figsize=(20,15))
ax = plt.subplot(111)
plt.hist(data1['z_spec'], color='red', bins=np.arange(0.33,0.37,0.0025), range=None, histtype='bar', align='mid',rwidth=.95, label='possible members')

#plot a new subset of candidates that is most likely the actual cluster members
plt.hist(data2['z_spec'], color='darkorange', bins=np.arange(0.33,0.37,0.0025), range=None, histtype='bar', align='mid',rwidth=.95, label='cluster members')

#plot mean and 3 sigma of those probable cluster members
mean=stat.mean(data2['z_spec'])
stdev=stat.stdev(data2['z_spec'])
plt.axvline(x=mean, color='darkslategrey', linewidth=3, label='mean')
plt.axvline(x=mean+(3*stdev), color='deepskyblue', linewidth=3, label='3 sigma')
plt.axvline(x=mean-(3*stdev), color='deepskyblue', linewidth=3)

#labels
plt.xlabel('redshift', fontsize=20)
plt.xticks(np.arange(0.33,0.37,0.0025), fontsize=15)
plt.yticks(np.arange(0,11,1), fontsize=15)
plt.title('Cluster Members per Redshift', fontsize=20)
legend = ax.legend(loc='upper left',fontsize='x-large')
```
