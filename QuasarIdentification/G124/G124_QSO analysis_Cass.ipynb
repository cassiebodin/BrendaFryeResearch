## Spectrum Analysis Code for QSO's

```python
import matplotlib
import matplotlib.pyplot as plt
fig = plt.figure()

import numpy as np  
from astropy.io import fits
from decimal import Decimal
matplotlib.rcParams.update({'font.size': 18}) 

#Read fits file
hdul = fits.open('spec_QSO1.fits', memmap=True)  
# hdul.info()
data = hdul[1].data
cols = hdul[1].columns
#cols.info() # prints out the names of the columns
#print(data[0]) # prints out the first row of data (first element of each of column)

# Extract the wavelength (in units of "log Angstroms"), and flux (in units of "10e-17 erg/s/cm^2/Angstrom") 
# Call the wavelength "wl" and the flux "flux".  
wavetmp = data.field('loglam')
wl = 10**wavetmp  
flux = data.field('flux') 
                     
# Plot up the QSO spectrum. 
# call the plot parameter of "drawstyle='steps-mid'" to see exactly where the data points are.

#G124
z1=2.16 #cluster

z2=3.172016 #QSO1



#Ly_alpha

x1=1216.67*(1+z1)

x2=1216*(1+z2)

#CIV
x3_1=1548.203*(1+z1)
x3_2=1550.777*(1+z1)

x4_1=1548.203*(1+z2)
x4_2=1550.777*(1+z2)

#MgII
x5_1=2796.352*(1+z1)
x5_2=2803.531*(1+z1)

x6_1=2796.352*(1+z2)
x6_2=2803.531*(1+z2)

#Ca2K
x7_1=3934*(1+z1)
x7_2=3934*(1+z2)

#CaH
x8_1=3968*(1+z1)
x8_2=3968*(1+z2)

#plot the spectra
fig, ax = plt.subplots(figsize=(16,8))
ax.plot(wl, flux,drawstyle='steps-mid')

#LyA
plt.axvline(x1, color='red')
plt.text(x1,8,'LyA')

plt.axvline(x2, color ='green')
plt.text(x2,12,'LyA')

#CIV
plt.axvline(x3_1, color='red')
plt.text(x3_1,9,'CIV_1')
plt.axvline(x3_2, color='red')
plt.text(x3_2,8,'CIV_2')

plt.axvline(x4_1, color ='green')
plt.text(x4_1,7,'CIV_1')
plt.axvline(x4_2, color ='green')
plt.text(x4_2,6,'CIV_2')

#MGII
plt.axvline(x5_1, color='red')
plt.text(x5_1,11,'MgII_1')
plt.axvline(x5_2, color='red')
plt.text(x5_2,10,'MgII_2')

# MgII, Ca2K, CaH are both out of range

#plt.axhline(y=12,color='purple')
ax.set(xlabel='Observed Wavelength (Angstroms)', xscale = 'linear', ylabel='Flux (x 10^-17 erg/s/cm^2/Ang)', yscale = 'linear')

#Full spec
ax.set_xlim(3800, 10350) 
ax.set_ylim(-2,16)
#fig.savefig('QSO1.png') 

#Ly_alpha #comment out all of the plt lines for everything but ly_alpha
#ax.set_xlim(3800, 3900)
#ax.set_ylim(-2,10)
#fig.savefig('QSO1_zoom_lyalpha.png')

#CIV #comment out all of the plt lines for everything but CIV
#ax.set_xlim(4800, 5000) 
#ax.set_ylim(0,10)
#fig.savefig('QSO1_zoom_civ.png') 

#MgII #comment out all of the plt lines for everything but MgII
#ax.set_xlim(8750, 8950) 
#ax.set_ylim(-2,8)
#fig.savefig('QSO1_zoom_mgii.png')
```
